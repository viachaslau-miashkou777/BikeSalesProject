## 5. Sales costs, profit, and revenue:

```python
# Read the dataset from the CSV file
df_sales = pd.read_csv("bike_sales.csv")

# Sales costs, profit, and revenue by year
df_sales_year = df_sales.groupby("Year").agg({"Unit_Cost":"sum", "Profit":"sum", "Revenue":"sum"})

# Annual plot
plt.figure(figsize=(8,6))
plt.plot(df_sales_year.index, df_sales_year["Unit_Cost"], label="Cost")
plt.plot(df_sales_year.index, df_sales_year["Profit"], label="Profit")
plt.plot(df_sales_year.index, df_sales_year["Revenue"], label="Revenue")
plt.xlabel("Year")
plt.ylabel("Amount ($)")
plt.title("Annual Costs, Profit, and Revenue")
plt.legend()
plt.show()
```

```python
# Monthly plot
# definning the order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

for year in range(2011, 2017):
    df_yearly = df_sales[df_sales['Year'] == year].copy()
    
    # convert 'Month' column to categorical with correct ordering
    df_yearly['Month'] = pd.Categorical(df_yearly['Month'], categories=month_order, ordered=True)

    
    df_monthly = df_yearly.groupby('Month')[['Unit_Cost', 'Profit', 'Revenue']].sum()
    plt.figure(figsize=(12,6))
    plt.plot(df_monthly.index, df_monthly['Unit_Cost'], label='Cost')
    plt.plot(df_monthly.index, df_monthly['Profit'], label='Profit')
    plt.plot(df_monthly.index, df_monthly['Revenue'], label='Revenue')
    plt.xlabel('Month')
    plt.ylabel('Amount ($)')
    plt.title(f'Monthly Costs, Profit, and Revenue for {year}')
    plt.legend()
    plt.show()
```

## 6. Customers by Age Group

```python
# Customers by age group by year
df_age = df_sales.groupby(["Year","Age_Group"]).size().unstack()
df_age.plot(kind="bar", stacked=True, figsize=(12,6))
plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.title("Bike Store Customers by Age Group by Year")
plt.show()
```

```python
# Sales costs, profit, and revenue by year by age group
df_age_sales = df_sales.groupby(["Year","Age_Group"]).agg({"Unit_Cost":"sum", "Profit":"sum", "Revenue":"sum"}).reset_index()
df_age_sales.pivot(index="Year", columns="Age_Group", values=["Unit_Cost","Profit","Revenue"]).plot(kind="bar", stacked=True, figsize=(12,6))
plt.xlabel("Year")
plt.ylabel("Amount ($)")
plt.title("Unit Costs, Profit, and Revenue by Year by Age Group")
plt.show()
```

## 7. Customers by Gender

```python
# Number of customers by gender by year
df_gender_number = df_sales.groupby(["Year","Customer_Gender"]).size().unstack()
df_gender_number.plot(kind="bar", stacked=True, figsize=(12,6))
plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.title("Number of Customers by Gender by Year")
plt.show()
```

```python
# Sales costs, profit, and revenue by year and by gender
df_gender_sales = df_sales.groupby(["Year","Customer_Gender"]).agg({"Unit_Cost":"sum", "Profit":"sum", "Revenue":"sum"}).reset_index()
df_gender_sales.pivot(index="Year", columns="Customer_Gender", values=["Unit_Cost","Profit","Revenue"]).plot(kind="bar", stacked=True, figsize=(12,6))
plt.xlabel("Year")
plt.ylabel("Amount ($)")
plt.title("Unit Costs, Profit, and Revenue by Year and by Gender")
plt.show()
```

## 8. Customers by Country

```python
# Number of customers by country and year
df_country = df_sales.groupby(["Year","Country"]).size().unstack()
df_country.plot(kind="bar", stacked=True, figsize=(12,6))
plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.title("Number of Customers by Country and Year")
plt.show()
```

```python
# Sales costs, profit, and revenue by country and year
df_gender_sales = df_sales.groupby(["Year","Country"]).agg({"Unit_Cost":"sum", "Profit":"sum", "Revenue":"sum"}).reset_index()
df_gender_sales.pivot(index="Year", columns="Country", values=["Unit_Cost","Profit","Revenue"]).plot(kind="bar", stacked=True, figsize=(12,6))
plt.xlabel("Year")
plt.ylabel("Amount ($)")
plt.title("Unit Costs, Profit, and Revenue by Country and Year")
plt.show()
```
